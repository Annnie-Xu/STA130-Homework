An attribute refers to the variables or data of an object. A method is a function that can be used to modify the data of an object or a class.

Count: The number of non-missing values in the column. This count does not include NaN (not a number) values.

Mean: The average of the values in the column. It is calculated by summing all the values and dividing by the count of non-missing values.

Std (Standard Deviation): A measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean, while a high standard deviation indicates that the values are spread out over a wider range.

Min (Minimum): The smallest value in the column.

25% (25th Percentile): The value below which 25% of the data falls. It is also known as the first quartile.

50% (50th Percentile or Median): The value below which 50% of the data falls. It is the middle value of the dataset, and if the number of observations is even, it is the average of the two middle numbers.

75% (75th Percentile): The value below which 75% of the data falls. It is also known as the third quartile.

Max (Maximum): The largest value in the column.

(a) an example of a "use case" in which using df.dropna() might be peferred over using del df['col']:
    When we have a dataset with a few columns where missing values are sporadic but crucial, for example, a dataset tracking stock prices where certain dates have missing values. In this case, df.dropna() might be preferred as we would want to remove rows where any column has missing data.
    
(b) When we have a dataset with a column having a significant amount of missing data compared to other columns. For example, if there is a ColumnX which is not critical but has a lot of missing data, removing this column using del df['col'] might be preferred.

(c) When using both methods together, applying del df['col'] before df.dropna() can be important because removing columns with too many missing values first can reduce the number of rows that df.dropna() might need to process. Furthermore, by dropping irrelevant or excessively missing columns first, we will be able to ensure that df.dropna() focuses only on the remaining data.


```python
# (d) 
# Display basic info about the dataset before cleaning
print("Before cleaning:")
print(df.info())
print("\nMissing data before cleaning:")
print(df.isnull().sum())

#Example: Removing a column with significant missing data
column_to_remove = 'Estimated Shares Outstanding'
if column_to_remove in df.columns:
    del df[column_to_remove]

# Alternatively, drop rows with any missing values
df_cleaned = df.dropna()

# Display basic info about the dataset after cleaning
print("\nAfter cleaning:")
print(df_cleaned.info())
print("\nMissing data after cleaning:")
print(df_cleaned.isnull().sum())
```

Justification of approach: I use del to remove the column 'Estimated Shares Outstanding' because it has a lot of missing values and is not that relevant to the analysis. After removing irrelevant columns, I use df.dropna() to remove rows with any missing values in the remaining columns.

1.First, df.groupby("col1")["col2"].describe() will classify and group "col2" according to "col1". Then, the function will describe and calculate the count, mean, standard deviation, minimum, Q1, median, Q3, and maximum value of "col2". For example, in the dataset I'm using, let's take "Ticker Symbol" as 'col1' and "Cash Ratio" as 'col2'. df.groupby('Ticker Symbol')['Cash Ratio'].describe() will get me descriptive statistics for the Cash Ratio column, grouped by Ticker Symbol. 2.for df.describe(), the count shows how many valid entries exist in each column across the entire dataset. Missing values in any column will reduce this count compared to the total number of rows. On the other hand, for df.groupby("col1")["col2"].describe(), the count shows how many valid entries exist in col2 for each subgroup defined by col1. This can vary depending on how many valid entries of col2 are present within each group. If a group has missing values in col2, its count will be lower, but this count is specific to that subgroup. df.describe() provides an overall picture of the dataset’s completeness for each column, while df.groupby("col1")["col2"].describe() provides insights into the completeness of col2 within each subgroup defined by col1, which can vary depending on how the data is grouped.
